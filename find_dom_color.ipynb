```python
def get_dom_color(dir_path):
    print(dir_path)
    try:
        color_thief = ColorThief(dir_path + '/out.jpeg')
        dom_col = "rgb" + str(color_thief.get_color(quality=1))
        return(dom_col)
    except: 
        return('rgb(0,0,0)')
```

```python
get_dom_color(directs[2])
```

```python
colors = [get_dom_color(county) for county in directs]
```

```python
import csv
from itertools import izip

with open('colors_by_county.csv', 'wb') as f:
    writer = csv.writer(f)
    writer.writerows(izip(directs, colors))
```
